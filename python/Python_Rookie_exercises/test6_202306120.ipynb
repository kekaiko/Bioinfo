# Counting Point Mutations

##Problem
Given two strings s  and t  of equal length, the Hamming distance between s  and t , denoted dH(s,t) , is the number of corresponding symbols that differ in
s and t






### Sample Dataset
GAGCCTACTAACGGGAT
CATCGTAATGACGGCCT

```python

#Counting Point Mutations
seq1="GAGCCTACTAACGGGAT"
seq2="CATCGTAATGACGGCCT"

Hamming_distance=0
#首先判断seq1和seq2的长度是否相等
if len(seq1)==len(seq2):
    #如果相等，就逐个比较两个序列的碱基是否相同
    for i in range(len(seq1)):
        #如果不同，Hamming_distance加1
        if seq1[i]!=seq2[i]:
            Hamming_distance+=1
        else:
            continue
    print(Hamming_distance)
else:
    print("The length of seq1 and seq2 is not equal!")

```
